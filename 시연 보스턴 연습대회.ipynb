```python
import pandas as pd
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge

import warnings
warnings.filterwarnings("ignore")
```

```python
train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Boston Housing Data/train.csv')
test= pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Boston Housing Data/test.csv')
submission = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/data/Boston Housing Data/sample_submission.csv')

print(train.head())
```

```python
X= train.drop(columns=['ID','MEDV'])
test= test.drop(columns=['ID'])
#데이터셋 정보 확인
X.info()
#결측치점검
X.isnull().sum() 
#통계값
X.describe()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

#정규분포화 (var=1)

train_scaler = scaler.fit_transform(train[train.columns[1:-1]])
test_scaler = scaler.transform(test)

test_scaler
```

```python
#ridge 규제 정도 설정한 모델 생성
model = Ridge(alpha =1.0)
model.fit(train_scaler, train['MEDV'])
#test_scaler예측하기
pred = model.predict(test_scaler)
pred
```
